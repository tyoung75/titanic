```python
import pandas as pd

train_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
train = pd.read_csv(train_url)

test_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv"
test = pd.read_csv(test_url)

#print head of test and train
print(train.head())
print(test.head())
```

```python
train_describe = train.describe()
train_shape = train.shape

print(train_describe)
print(train_shape)

test_describe = test.describe()
test_shape = test.shape

print(test_describe)
print(test_shape)
```

```python
# The number of people who survived (1) vs who died (0)
print(train["Survived"].value_counts())
# as a proportion
print(train["Survived"].value_counts(normalize = True))

# The number of men who survived vs who died
print(train["Survived"][train["Sex"] == 'male'].value_counts())
# The number of women who survived vs who died
print(train["Survived"][train["Sex"] == 'female'].value_counts())

# Proportion of men who died
print(train["Survived"][train["Sex"] == 'male'].value_counts(normalize = True))
# Propoertion of Women who died
print(train["Survived"][train["Sex"] == 'female'].value_counts(normalize = True))
```

```python
# Creating Child column
train["Child"] = float('NaN')

# Assigning 1 to passengers under 18, 0 to 18+.
train["Child"][train["Age"] < 18] = 1
train["Child"][train["Age"] >= 18] = 0

print(train["Child"])

#Print proportion of Survival Rates < 18
print(train["Survived"][train["Child"] == 1].value_counts(normalize = True))

#Print proportion of Survival Rates 18+
print(train["Survived"][train["Child"] == 0].value_counts(normalize = True))
```

```python
# Create a test copy
test_one = test

# Survived = 0
test_one["Survived"] = 0

# Set women in Surviveed to 1
test_one["Survived"][test_one["Sex"] == 'female'] = 1

print(test_one["Survived"])
```

```python
# DECISION TREE
import numpy as np
from sklearn import tree
```

```python
# Turning male (0) and female (1) groups into integer form
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1

#Impute Embarked variable
train["Embarked"] = train["Embarked"].fillna("S")

#Convert Embarked class to integer
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2

print(train["Sex"])
print(train["Embarked"])
```

```python
print(train)

#Change NA in Age to median
train["Age"] = train["Age"].fillna(train["Age"].median())

print(train["Age"].median())

#Create the target and features numpy arrays
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values

#Create decision tree
titanic_one = tree.DecisionTreeClassifier()
titanic_one = titanic_one.fit(features_one, target)

#Importance and score
print(titanic_one.feature_importances_)
print(titanic_one.score(features_one, target))
```

```python
#CLEANING TEST
#Change NA in Age to median
test["Age"] = test["Age"].fillna(test["Age"].median())

print(test_one)
# Turning male (0) and female (1) groups into integer form
test_one["Sex"][test_one["Sex"] == "male"] = 0
test_one["Sex"][test_one["Sex"] == "female"] = 1

test.Fare[152] = test.Fare.median()

#extract features from the test set
test_features = test[["Pclass", "Sex", "Age", "Fare"]].values

#Make predictions using the test set
prediction = titanic_one.predict(test_features)
print(prediction)

# Create a data frame with two columns: PassengerId & Survived. Survived contains your predictions
PassengerId =np.array(test["PassengerId"]).astype(int)
titanic_solution = pd.DataFrame(prediction, PassengerId, columns = ["Survived"])
print(titanic_solution)

# Check that your data frame has 418 entries
print(titanic_solution.shape)

# Write your solution to a csv file with the name my_solution.csv
titanic_solution.to_csv("titanic_solution_one.csv", index_label = ["PassengerId"])
```

```python
###### RandomForestClassifier ###
from sklearn.ensemble import RandomForestClassifier

# Take forest features wanted
features_forest = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values

#building and fitting my_forrest
forest = RandomForestClassifier(max_depth = 10, min_samples_split = 2, n_estimators = 100, random_state = 1)
my_forest = forest.fit(features_forest,target)

print(my_forest.score(features_forest,target))

#compute predictions on test set features then print the length of the prediction vector
test_features = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
pred_forest = my_forest.predict(test_features)
print(len(pred_forest))
```
